12. Добавьте колонку “Companion” с количеством попутчиков в одной каюте с этим
пассажиром.

```python
cabins = data['cabin'].value_counts()
cabins
```

```python
def choose_cabin(x):
    if x in cabins:
        return cabins[x] - 1
```

```python
data['companion'] = data['cabin'].apply(choose_cabin)
```

16. Отобразите сколько женщин и мужчин выжило/не выжило.

```python
surv_m = data[(data.sex == 'm') & (data.survived == 1)].shape[0]
nosurv_m = data[(data.sex == 'm') & (data.survived == 0)].shape[0]
surv_f = data[(data.sex == 'f') & (data.survived == 1)].shape[0]
nosurv_f = data[(data.sex == 'f') & (data.survived == 0)].shape[0]
```

```python
print(f'Мужчин выжило: {surv_m}',
      f'Мужчин не выжило: {nosurv_m}', sep=f'\n')
```

```python
print(f'Женщин выжило: {surv_f}',
      f'Женщин не выжило: {nosurv_f}', sep=f'\n')
```
